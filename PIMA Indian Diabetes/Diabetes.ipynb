### **Handling Outliers**

```python
#Quantiles of 25% and 75%
Q1 = df2['Insulin'].quantile(0.25)                  
Q3 = df2['Insulin'].quantile(0.75)

#Inter quartile range
IQR = Q3 - Q1

#Lower extremes and higher extremes
i_upper = (Q3+1.5*IQR)
i_lower = (Q1-1.5*IQR)

```

Only **Insulin** feature outliers gone far away 

See how will I cut them off

```python
#Checking how far they, just to be accurate
print(df2.Insulin.median())
print(df2.Insulin.max())
```

See nthe variation, generally maximum insulin level is nearly 170 for 2-Hour serum insulin (U/ml).
So considering this as data collection error and dropping the outliers

```python
#Dropping extreme outliers only 
outliers = df2[(df2['Insulin']<i_lower) | (df2['Insulin']>i_upper)].index
df2.drop(outliers,inplace = True)
```

## **Additional Analysis**

### ***Feature Scaling***

We are choosing normalisation techinique due to outliers in features

```python
from sklearn.preprocessing import MinMaxScaler
X = df2.drop(columns='Outcome')
norm = MinMaxScaler()

X = norm.fit_transform(X)
X
```

```python
from sklearn.decomposition import PCA

pca=PCA()
fit=pca.fit(X)
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.figure(figsize = (10,4))
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot of Elbow Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]

plt.figure(figsize = (15,4))
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

**Why is diabetes a challenge?**

In low- and middle-income countries (LMICs), the burden of non-communicable diseases such as diabetes is rapidly rising, overpassing the existing burden of communicable diseases. Patients with diabetes living in low-income communities face unique challenges related to lack of awareness, difficulty in accessing health care systems and medications, and consequently failure in achieving optimal diabetes management and preventing complications. Effective diabetes prevention and care models could help reduce the rising burden by standardizing guidelines for prevention and management, improving access to care, engaging community and peers, improving the training of professionals and patients and using the newest technology in the management of the disease
